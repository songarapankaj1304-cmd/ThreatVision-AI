# Visualization Notebook

This notebook is dedicated to visualizing the results of the ThreatVision AI project. It includes various plots and charts to help understand the data and model performance.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('../data/cicids_combined_data.csv')

# Display the first few rows of the dataset
data.head()
```

```python
# Visualize the distribution of attacks
plt.figure(figsize=(10, 6))
sns.countplot(data['attack_type'])
plt.title('Distribution of Attack Types')
plt.xlabel('Attack Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../outputs/charts/attack_distribution.png')
plt.show()
```

```python
# Visualize the correlation heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig('../outputs/charts/correlation_heatmap.png')
plt.show()
```
